## **4. Model Building:**

### **A. Split data into X and Y.**

```python
# Split data into X
# 'Class' variable is the Target Variable (y) and remaining are independent variables (x1,x2,...)
X = patient_data.drop(labels= "Class" , axis = 1)
X.head()
```

```python
# Create Target Variable (y) of the variable 'Class' from the dataframe
y = patient_data['Class']
y
```

### **B. Split data into train and test with 80:20 proportion.**

```python
# Import Library to Split Data
from sklearn.model_selection import train_test_split

# Split X and y into training and test set in 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
print('Shape of X_train',X_train.shape)
print('Shape of y_train',y_train.shape)
print('Shape of X_test',X_test.shape)
print('Shape of y_test',y_test.shape)
```

### **C. Train a Supervised Learning Classification base model using KNN classifier.**

```python
# Import KNN Classifier from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Instantiate learning model
knn1 = KNeighborsClassifier()

# fitting the model
knn1.fit(X_train, y_train)
```

### **D. Print all the possible performance metrics for both train and test data.**

```python
# Predict the Test Dataset
y_pred = knn1.predict(X_test)
y_pred
```

```python
# Print Train Data Performance
knn1.score(X_train,y_train)
```

```python
# Print Test Data Performance
knn1.score(X_test,y_test)
```

```python
# Predict from Training Dataset
y_pred_train = knn1.predict(X_train)
y_pred_train
```

```python
# Confusion matrix for Train Data
from sklearn import metrics

print("Confusion Matrix for Training Data")
cm=metrics.confusion_matrix(y_train, y_pred_train, labels=["Normal", "Type_H", "Type_S"])

df_cm = pd.DataFrame(cm, index = [i for i in ["Normal", "Type_H", "Type_S"]],
                  columns = [i for i in ["Normal", "Type_H", "Type_S"]])

import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize = (7,5));
sns.heatmap(df_cm, annot=True);
```

```python
# Confusion matrix for Test Data

print("Confusion Matrix for Test Data")
cm=metrics.confusion_matrix(y_test, y_pred, labels=["Normal", "Type_H", "Type_S"])

df_cm = pd.DataFrame(cm, index = [i for i in ["Normal", "Type_H", "Type_S"]],
                  columns = [i for i in ["Normal", "Type_H", "Type_S"]])

plt.figure(figsize = (7,5));
sns.heatmap(df_cm, annot=True);
```

```python
# Calculate performance metrics for training data
# Import inbuilt classification report from sklearn
from sklearn.metrics import classification_report
target_names = ['Normal','Type_H','Type_S']
print(classification_report(y_train, y_pred_train, target_names=target_names))
```

```python
# Calculate performance metrics test data
# Import inbuilt classification report from sklearn
print(classification_report(y_test, y_pred, target_names=target_names))
```

## **5. Performance Improvement:**

### **A. Experiment with various parameters to improve performance of the base model.**

```python
# Check for optimal value of k
kscores = []
for i in range(1,25):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(X_train, y_train)
  s = knn.score(X_test, y_test)
  kscores.append(s)
plt.plot(range(1,25), kscores, 'o-')   
plt.xticks(range(1,25))
plt.show()
```

*For k=1, the accuracy score is the highest. We build a model with k=1*

```python
# Predict the Test Dataset
y_pred2 = knn2.predict(X_test)
y_pred2
```

```python
# Print Test Data Performance
knn2.score(X_test,y_test)
```

```python
# Change weights parameter in the above model
knn3 = KNeighborsClassifier(n_neighbors = 1, weights='distance')

# fitting the model
knn3.fit(X_train, y_train)
# Predict the Test Dataset
y_pred3 = knn3.predict(X_test)
# Print Train Data Performance
print("Accuracy on Training Data:",knn3.score(X_train,y_train))
# Print Test Data Performance
print("Accuracy on Test Data:",knn3.score(X_test,y_test))
# Calculate performance metrics
print(classification_report(y_test, y_pred3, target_names=target_names))
```

*Not much improvement is seen by introducing the weights parameter.*

## **3. Data Preparation and model building:**

### **A. Split data into X and Y.**

[Recommended to drop ID & ZipCode. LoanOnCard is target Variable]

```python
# Drop ID, ZipCode and LoanOnCard from Independent Variables
X = bank_data.drop(['ID','ZipCode','LoanOnCard'], axis=1)
# Dispaly Independent Variables
X.head()
```

```python
# Drop all variables except Target Variable LoanOnCard
Y = bank_data['LoanOnCard']
print(Y)
```

### **B. Split data into train and test. Keep 25% data reserved for testing.**

```python
# Create copy of test data
xtestsc = x_test

# Apply Z-score scaling to numerical columns of test data
xtestsc[col_names] = stats.zscore(xtestsc[col_names])
xtestsc.head()
```

```python
# Tuning KNN Model
# Check for optimal value of k
kscores = []
for i in range(1,25):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(xtrainsc, y_train)
  s = knn.score(xtestsc, y_test)
  kscores.append(s)
plt.plot(range(1,25), kscores, 'o-')   
plt.xticks(range(1,25))
plt.show()
```
